```python
import os
import zipfile
```

```python
# Specify the input folder containing the TIFF files and the output ZIP file
input_folder = 'Output'
output_zip = 'Output/Alice_Arm.zip'
```

```python
# Ensure the output directory for the ZIP file exists
os.makedirs(os.path.dirname(output_zip), exist_ok=True)
```

```python
# Initialize a flag to track if files are added
files_added = False
```

```python
# Create the ZIP file and add all .tif/.tiff files from the input folder
with zipfile.ZipFile(output_zip, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for file in os.listdir(input_folder):  # Only scan the specific folder
        if file.lower().endswith(('.tif', '.tiff')):  # Check for .tif or .tiff
            file_path = os.path.join(input_folder, file)  # Full file path
            zipf.write(file_path, file)  # Add file to ZIP
            files_added = True
            print(f"Added to ZIP: {file_path}")

# Check if any files were added
if files_added:
    print(f"TIFF files successfully zipped into: {output_zip}")
else:
    print(f"No TIFF files found in the directory: {input_folder}")
```
